```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the graph from the file
G = nx.read_edgelist('soc-Epinions1.txt', create_using=nx.DiGraph())

# Print basic information about the graph
print("Number of nodes:", G.number_of_nodes()) # 75879
print("Number of edges:", G.number_of_edges()) # 508837
```

```python
# Compute in-degree and out-degree distributions
in_degrees = dict(G.in_degree())
out_degrees = dict(G.out_degree())

# Calculate the maximum degree
max_degree = max(max(in_degrees.values()), max(out_degrees.values()))

# Compute the frequency of each degree
in_degree_freq = np.zeros(max_degree + 1)
out_degree_freq = np.zeros(max_degree + 1)

for degree in in_degrees.values():
    in_degree_freq[degree] += 1

for degree in out_degrees.values():
    out_degree_freq[degree] += 1

# Normalize the frequency
in_degree_prob = in_degree_freq / sum(in_degree_freq)
out_degree_prob = out_degree_freq / sum(out_degree_freq)

# Plot the in-degree distribution
plt.figure(figsize=(10, 5))
plt.plot(range(len(in_degree_prob)), in_degree_prob, 'bo', markersize=8)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('In-degree')
plt.ylabel('Probability')
plt.title('In-degree Distribution')
plt.grid(True)
plt.show()

# Plot the out-degree distribution
plt.figure(figsize=(10, 5))
plt.plot(range(len(out_degree_prob)), out_degree_prob, 'ro', markersize=8)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Out-degree')
plt.ylabel('Probability')
plt.title('Out-degree Distribution')
plt.grid(True)
plt.show()
```
